```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = stopwords.words('english')

```

```python
def clean_sentence(sen):
    sen = re.sub("\d+|\.|,|Md$|%|:|$|\)|\(|\n", '', sen).lower()

    senList = word_tokenize(sen.lower())
    DocFil = [w for w in senList if not w in stopwords.words('english')]
    senFil = ' '.join(DocFil)
    return senFil
```

```python
vectorizer = TfidfVectorizer()
```

```python
Data.abstract.values # abstract column of the data 
```

```python
list(Data.abstract.values)
```

```python
abstractsList = [clean_sentence(abs) for abs in list(Data.abstract.values)]
abstractsList #abstract= [abs1,abs2,abs3]
```

```python
vectors = vectorizer.fit_transform(abstractsList)
feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
df
```

```python
query = "education best system africa"
```

```python
engine = df.transpose()

colnames = ["Chapitre_"+str(i+1) for i in range(20)]
engine.columns = colnames
#engine.head()
```

```python
query_data = engine.loc[word_tokenize(clean_sentence(query))]
tfidf = query_data.sum()
tfidf.sort_values(ascending = False).head(10)
```

```python
def book_engine(query, engine):
    word_tokenize(clean_sentence(query))
    query_data = engine.loc[word_tokenize(clean_sentence(query))]
    tfidf = query_data.sum()
    return(tfidf.sort_values(ascending = False).head(10))
```

```python
query = "music technology education Africa"
engine = engine
book_engine(query, engine)
```
